```python
import cv2 as cv
from matplotlib.pyplot import imshow, show, figure
import imutils
import numpy as np

def cv2_imshow(image, figsize=(3,3)):
    fig = figure(figsize=figsize)
    try:
        imshow(image[:,:,::-1])
    except IndexError:
        imshow(image, cmap="gray")
    show()
```

```python

def extract_corners(filename):
    kernel = np.ones((3,3),np.uint8)
    #Read original image and grayscale
    img_original = cv.imread(filename)
    img = cv.cvtColor(cv.imread(filename), cv.COLOR_BGR2GRAY)
    img_copy = img.copy()
    #cv2_imshow(img_original)
    #Transform in black and white
    imgt = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY_INV,57,3)
    #cv2_imshow(imgt)
    imgt_copy = imgt.copy()
    #Find external countour, create mask and draw external countour on mask
    contours, hierarchy = cv.findContours(image=imgt_copy, mode=cv.RETR_EXTERNAL, method=cv.CHAIN_APPROX_SIMPLE)
    mask = np.zeros_like(imgt)
    max_c = 0
    max_area = 0
    for contorno in contours:
        c_area = cv.contourArea(contorno)
        if c_area > max_area:
            max_area = c_area
            max_c = contorno
    cv.drawContours(image=mask, contours=[max_c], contourIdx=-1, color=(255), thickness=2, lineType=cv.LINE_AA)
    #Find grid corners and draw on original image
    #cv2_imshow(mask)
    corners = cv.goodFeaturesToTrack(mask, 4, 0.4, 50)
    for corner in corners:
        x,y = corner.ravel().astype(int)
        cv.circle(img_original, (x,y), 5, (36,255,12), -1)
    corners = sorted(corners.copy().astype(np.float32).reshape(corners.shape[0], corners.shape[2]), key=lambda x: x[0])
    pts1 = []
    a,b,c,d = corners
    if a[1]>b[1]:
        pts1.append(list(a))
        pts1.append(list(b))
    else:
        pts1.append(list(b))
        pts1.append(list(a))

    if c[1]>d[1]:
        pts1.append(list(c))
        pts1.append(list(d))
    else:
        pts1.append(list(d))
        pts1.append(list(c))

    # pts1 = Bot-Left; Top-Left; Bot-Right; Top-Right
    pts1 = np.array(pts1)
    return pts1
```

```python
pts1 = extract_corners("sudoku_digital.jpg")
```

```python
# Apply perspective transformation
def perspective_transformation(filename, imsize, pts1):
    img_clean = cv.imread(filename)
    pts2 = np.float32([[0,imsize], [0,0], [imsize,imsize], [imsize,0]])
    for val in pts1:
        cv.circle(img_clean,(int(val[0]),int(val[1])),1,(0,255,0),-1)
    M = cv.getPerspectiveTransform(pts1,pts2)
    dst = cv.warpPerspective(img_clean,M,(imsize,imsize))
    return dst
```

```python
transformed_img = perspective_transformation("sudoku_digital.jpg", 270, pts1) #imgsize should be a multiple of 9
cv2_imshow(transformed_img)
```

```python
imgt = cv.adaptiveThreshold(cv.cvtColor(transformed_img, cv.COLOR_BGR2GRAY),255,cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY_INV,31,30)
d = cv.dilate(imgt, (3,3))
p = cv.erode(d, (3,3))
# p = cv.morphologyEx(imgt, cv.MORPH_CLOSE, (5,5))
cv2_imshow(p)
```

```python
def show_all_cells(img):
    h, w = img.shape
    cell_size = int(h/9)
    min_x = 0
    min_y = 0
    for i in range(9):
        for j in range(9):
            cell_img = img[min_x:(min_x+cell_size), min_y:(min_y + cell_size)]
            cv2_imshow(cell_img)
            min_x += cell_size
        min_x = 0
        min_y += cell_size
```

```python
show_all_cells(p)
```
